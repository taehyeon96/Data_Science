### * Plotting iris dataset by using Pandas

```python
import pandas as pd
```

```python
data = load_iris()

# 데이터를 DataFame으로 그냥 불러오면 이렇게 된다.
pd.DataFrame(data.data)
```

```python
# 앞서 분석했던 "dir을 확인하여" columns를 알맞게 불러오기 <중요>
iris = pd.DataFrame(data.data, columns=data.feature_names)

pd.DataFrame(iris)
```

```python
# 둘의 차이 확인하기

# Pandas 미사용 출력
data.target

# Pandas 사용 출력
iris['target'] = data.target
pd.DataFrame(iris['target'])
```

#### * Box-Plotting of iris Dataset
    - Box plot : 많은 데이터를 눈으로 확인하기 어려울 때 
                데이터 범위, 중앙값 등을 확인하기 용이
                

      
      
 * IQR = Q3 - Q1

```python
# sklean의 boxplot()을 활용
iris.iloc[:,:-1].boxplot(figsize=(10,5))
```

```python
# pandas의 plotting을 활용
pd.plotting.scatter_matrix(iris, figsize=(10,10));
# 아웃풋 안나오게 하려면 ; 붙인다.
```

#### * 왜도(skewness)와 첨도(kurtosis)
    - 왜도 : 분포가 정규분포에 비해 얼마나 비대칭인지 나타내는 척도
    - 첨도 : 샘플의 점수가 평균을 중심으로 가까이 몰려 있을수록 분포의 정점은 더욱 뽀죡한 모양 즉, 첨도가 커짐

```python
iris.skew()
```

```python
iris.kurtosis()
```
